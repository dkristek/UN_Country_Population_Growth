```python
## Importing csv files as DF
import pandas as pd

country_pop_df = pd.read_csv("CountryPopulationForecastingResources/PopulationData.csv")

country_ref_df = pd.read_csv("CountryPopulationForecastingResources/CountriesData.csv")

global_pop_df = pd.read_csv("CountryPopulationForecastingResources/GlobalCountriesPopulation21-22.csv")
```

```python
# View pop df
country_pop_df.head(10)
```

```python
# View country code df
country_ref_df.head(10)
```

```python
# View global pop df
global_pop_df.head(10)
```

```python
# Merge DFs based on country
by_country_pop_df = country_pop_df.merge(country_ref_df, how="left", on="country_id")
by_country_pop_df.head(10)
```

```python
# population in 1000s to raw population number
by_country_pop_df["population"] = by_country_pop_df["population"]*1000
```

```python
by_country_pop_df.head(10)
```

```python
# global population in 1000s to raw population number
global_pop_df[["2021","2022"]] = global_pop_df[["2021", "2022"]] * 1000
global_pop_df.head(10)
```

```python
# Drop 2021 data
global_2022_pop_df = global_pop_df.drop(columns=["2021"])
global_2022_pop_df.head(10)
```

```python
# Get statistics
global_2022_pop_df["2022"].describe()
```

```python
# Order countries by population
global_2022_pop_df = global_2022_pop_df.sort_values(["2022"], ascending=False)
global_2022_pop_df.head(10)
```

```python
global_2022_pop_df.tail(20)
```

```python
# Create range bins and labels for countries by population
range_values = [1000, 100000, 1000000,5000000,10000000,30000000,250000000,1000000000,2000000000]

global_2022_pop_df["2022"].groupby(pd.cut(global_2022_pop_df["2022"], range_values)).count()
```

```python
#Categorizing population
g_names = ["Less Than 100,000", "100,000 - 1M", "1M - 5M", "5M - 10M", "10M - 30M", "30M - 250M", "250M - 1B", "Greater Than 1B"]

global_2022_pop_df["pop_cat"]=pd.cut(global_2022_pop_df["2022"], range_values, labels=g_names)

global_2022_pop_df.head(10)
```

```python
# Adding rank to DF 
global_2022_pop_df = global_2022_pop_df.reset_index()

global_2022_pop_df = global_2022_pop_df.rename(columns={"index":"Rank"})
global_2022_pop_df["Rank"] = global_2022_pop_df.index + 1

global_2022_pop_df.head()
```

```python
# Pull out top 4 country data 
top_4_df = global_2022_pop_df.loc[0:3, ['Country','2022']]
top_4_df 
```

```python
# append 2022 country to 1990 - 2021 country list
for index, row in top_4_df.iterrows():
    by_country_pop_df = by_country_pop_df.append({'mid_year':'2022', 
                              'population': row['2022'],
                              'country_name': row['Country']}, ignore_index=True)

# by_country_pop_df.drop(index=by_country_pop_df.index[-1], axis=0, inplace=True)
by_country_pop_df.tail(10)
```

```python
# Sort Df by country then year 
by_country_pop_df = by_country_pop_df.sort_values(['country_name', 'mid_year'], ascending=True)
by_country_pop_df.tail()
```

```python
# Drop Record key and country id
by_country_pop_df = by_country_pop_df.drop(columns=['record_key', 'country_id'])
by_country_pop_df.head(10)
```

```python
# Check dataypes
by_country_pop_df.dtypes
```

```python
# convert year and population to integers
by_country_pop_df['population'] = by_country_pop_df['population'].astype('int64')
```

```python
# Convert year to dt year
by_country_pop_df["mid_year"] = pd.to_datetime(by_country_pop_df["mid_year"], format="%Y")

by_country_pop_df.dtypes
```

```python
# View DF with converted DT
by_country_pop_df.head(10)
```

```python
# Export as new csv file
by_country_pop_df.to_csv("CountryPopulationForecastingResources/popsByCountries.csv", index=False)
```

```python
# Export 2022 global pop as new csv file
global_2022_pop_df.to_csv("CountryPopulationForecastingResources/globalPop2022.csv", index=False)
```
