```python
# import testing, errors, and predictions csv
import pandas as pd

#Model Testing 
china_test_df = pd.read_csv("Predictions/china_train_test.csv")

india_test_df = pd.read_csv("Predictions/india_train_test.csv")

america_test_df = pd.read_csv("Predictions/usa_train_test.csv")

indonesia_test_df = pd.read_csv("Predictions/indonesia_train_test.csv")

#Error Results of Model
china_error_df = pd.read_csv("Predictions/china_errors.csv")

india_error_df = pd.read_csv("Predictions/india_errors.csv")

america_error_df = pd.read_csv("Predictions/usa_errors.csv")

indonesia_error_df = pd.read_csv("Predictions/indonesia_errors.csv")

# Predictions
china_pred_df = pd.read_csv("Predictions/china_predictions.csv")

india_pred_df = pd.read_csv("Predictions/india_predictions.csv")

america_pred_df = pd.read_csv("Predictions/america_predictions.csv")

indonesia_pred_df = pd.read_csv("Predictions/indonesia_predictions.csv")
```

```python
# View each model df
print(china_test_df, india_test_df, america_test_df, indonesia_test_df)
```

```python
# View each results df
print(china_error_df, india_error_df, america_error_df, indonesia_error_df)
```

```python
# View each predictions df
print(china_pred_df, india_pred_df, america_pred_df, indonesia_pred_df)
```

```python
# Add country columns
# list to hold country name
countries = ["China", "India", "United States of America", "Indonesia"]

china_test_df["Country"] = countries[0]
china_error_df["Country"] = countries[0]
china_pred_df["Country"] = countries[0]

india_test_df["Country"] = countries[1]
india_error_df["Country"] = countries[1]
india_pred_df["Country"] = countries[1]

america_test_df["Country"] = countries[2]
america_error_df["Country"] = countries[2]
america_pred_df["Country"] = countries[2]

indonesia_test_df["Country"] = countries[3]
indonesia_error_df["Country"] = countries[3]
indonesia_pred_df["Country"] = countries[3]
    
```

```python
# Merge country df based on type

testing_df = pd.concat([china_test_df, india_test_df, america_test_df, indonesia_test_df], ignore_index = True)

error_df = pd.concat([china_error_df, india_error_df, america_error_df, indonesia_error_df], ignore_index = True)

predictions_df = pd.concat([china_pred_df, india_pred_df, america_pred_df, indonesia_pred_df], ignore_index = True)

print(testing_df, error_df, predictions_df)
```

```python
# Cleaning testing df - getting dtypes 
testing_df.columns = ["Year", "Predicted Pop", "Actual Pop", "Country"]

testing_df.dtypes
```

```python
# Converting to raw numbers (in thousands)
testing_df[["Predicted Pop", "Actual Pop"]] = testing_df[["Predicted Pop", "Actual Pop"]] * 1000

testing_df
```

```python
# Changing data types - year to DT and numbers to int64
testing_df["Year"] = pd.to_datetime(testing_df["Year"], format="%Y")

testing_df["Predicted Pop"] = testing_df["Predicted Pop"].astype("int64")

testing_df.dtypes
```

```python
# Cleaning predictions df - combining rows
predictions_df['0'] = predictions_df["0"].combine_first(predictions_df["predicted_mean"])
predictions_df
```

```python
# Renaming columns, drop column, getting dtype
predictions_df = predictions_df.drop(columns=['predicted_mean'])

predictions_df.columns=["Year", "Predicted Pop", "Country"]

predictions_df.dtypes
```

```python
# Converting to raw numbers and setting data type
predictions_df["Predicted Pop"] = (predictions_df["Predicted Pop"].astype("int64")) * 1000

predictions_df["Year"] = pd.to_datetime(predictions_df["Year"], format="%Y")

predictions_df.dtypes
```

```python
predictions_df
```

```python
# Cleaning error df - dropping empty column and formatting precentage

error_df = error_df[["MAPE","RMSE","Country"]] 

error_df['MAPE'] = error_df['MAPE'].map("{:,.2%}".format)

error_df
```

```python
# Exporting data as csv

testing_df.to_csv("Visualizations/ARIMA_testing.csv")

error_df.to_csv("Visualizations/ARIMA_testing_error.csv")

predictions_df.to_csv("Visualizations/ARIMA_predictions.csv")
```
